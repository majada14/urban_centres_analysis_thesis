```python
def dbscan_clustering(eps_value, el_value, dataset):
    # ---
    # compute dbscan clustering for specific subset with specified parameters
    # ---
    
    # initialize clustering
    dbscan = DBSCAN(eps=eps_value*1000,    # eps need to be settled in meters to fit the coordinates of data
                        min_samples=el_value,
                        algorithm='auto',
                        metric='euclidean')

    ids = dataset.node_id.values.astype(str)
    
    res = dbscan.fit(dataset[['lat_m', 'lon_m']]) # fit to the data directly
    
    clusters = res.labels_ # get the cluster labels
    num_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)  # to exclude the noise cluster from the counting
     
    cluster_df = pd.DataFrame({'id': ids, "clusters": clusters})
    cluster_df['id'] = cluster_df['id'].astype("string") 
    complete_subset = dataset.merge(cluster_df, left_on='node_id',right_on="id", how='left')
    
    noise_dataset = complete_subset[complete_subset.clusters ==-1] 
    cluster_dataset = complete_subset[complete_subset.clusters !=-1]
    
    return cluster_dataset, noise_dataset
    
```

```python
def polygon_construction(dataset):
    # ---
    # from subset of nodes data construct polygons figure containing the selected data inside
    # ---
    
    # --- data preparation
    gdf = gpd.GeoDataFrame(dataset, geometry='geometry')
    gdf.set_crs(epsg=32632, inplace=True)
    gdf['lat_m'] =gdf.geometry.y
    gdf['lon_m'] = gdf.geometry.x
    gdf['id'] = gdf['id'].astype(int)
    
    # --- polygons construction
    dissolved = gdf.dissolve(by='new_labels').copy()
    dissolved['geometry'] = dissolved.geometry.convex_hull
    dissolved['geometry'] = dissolved.geometry.buffer(50) # add a buffer to keep clusters on the same line

    convex_gdf = dissolved.reset_index()
    convex_gdf = convex_gdf[['new_labels', 'geometry' ]]
    
    # --- cleaning 
    final_polygons = convex_gdf[convex_gdf.geometry.type.isin(['Polygon', 'MultiPolygon'])]

    final_polygons_cleaned = final_polygons.copy()
    print(f"total urban centres identified: {final_polygons.shape[0]}") # sanity check
    
    # clusters that are almost completely (90%) inside other clusters are dropped -> important for the buffered versions
    final_polygons_cleaned["area"] = final_polygons_cleaned.area

    # overlay: we are checking for polygons that instersect with other polygons
    inter = final_polygons_cleaned.overlay(
        final_polygons_cleaned,
        how="intersection"
    )
    
    # removing self-intersections
    inter = inter[ inter["new_labels_1"] != inter["new_labels_2"] ]
    
    # then we compute the area (as fraction) of intersection beween each pair og polygons interstecting
    inter["frac"] = inter.area / inter.area_1
    
    # and we decide to drop the smallest polygon when the proportion of overaly is more than 0.95 
    to_drop = inter.loc[ inter["frac"] > 0.90, "new_labels_1" ].unique()

    # drop them
    final_polygons_cleaned = final_polygons_cleaned[~final_polygons_cleaned["new_labels"].isin(to_drop)]
    print(f"urban centres identified after cleaning process: {final_polygons_cleaned.shape[0]}")
    
    return final_polygons_cleaned
```
