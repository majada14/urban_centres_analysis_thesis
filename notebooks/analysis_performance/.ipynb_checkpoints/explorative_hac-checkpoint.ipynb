# functions setting

```python
distance_parameter = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0, 2.1, 2.2, 2.3, 2.4, 2.5]
```

```python
def dist_matrix_extraction(setting, polygon):
    
    current_matrix = pd.read_csv("datasets/matrix_computation/" + setting + "/polygon" + str(polygon) + ".csv", index_col=0)
    current_matrix.columns = current_matrix.columns.astype(int)
    
    return current_matrix
```

```python
def compound_validity_index(dataset, matrix):

    labels = dataset['clusters'].values
    
    ### ---
    
    # 0) Outlier ratio
    noise_ratio = np.sum(labels == -1) / len(labels)
    noise_norm = 1 - noise_ratio  # NORMALIZATION

    ### ---
    
    # 1) Cophenetic correlation coefficient
    matrix_condense = squareform(matrix)
    Z = linkage(matrix_condense, method='average', metric='precomputed')   # hierarchical clustering
    ccc, _ = cophenet(Z, matrix_condense)  
    ccc_norm = (ccc + 1) / 2  # NORMALIZATION
   
    
    ### ---
    # preparation: remove noise data since these metrics are computed on the clusters data!!
    clustered_data = dataset[dataset['clusters'] != -1].copy()
    clustered_ids = clustered_data['node_id'].values.astype(int)
    submatrix = matrix.loc[clustered_ids, clustered_ids].to_numpy()
    

    # 2) Silouhette score -> check the compactness of results: how much clusters are densely distributed and well separated from each other
    sil_score = silhouette_score(submatrix, clustered_data['clusters'].values, metric='precomputed')
    sil_norm = (sil_score + 1) / 2 # NORMALIZATION
     
    ### ---
    
    # 3) Davies-Boulin score -> measures for separation between clusters
    dbi_score = davies_bouldin_score(clustered_data[['lat_m', 'lon_m']].values, clustered_data['clusters'].values)
    dbi_norm = 1 - (dbi_score / (dbi_score + 1)) # NORMALIZATION

    # compound validity score
    cvs = np.mean([noise_norm, ccc_norm, sil_norm, dbi_norm])
    res = [round(noise_norm, 2), round(ccc_norm, 2), round(sil_norm, 2), round(dbi_norm, 2), round(cvs, 2)]
    
    return res
    
```

```python
def hac_clustering(dist, matrix, dataset, grid, setting):
    
    # initialize clustering
    hac = AgglomerativeClustering(
        n_clusters=None,               
        metric='precomputed',        
        linkage='average',             
        distance_threshold=dist)
    
    res = hac.fit(matrix) # fit to the distance matrix
    ids = [int(el) for el in matrix.index]
    
    clusters = res.labels_ # get the cluster labels
    
    cluster_df = pd.DataFrame({'id': ids, "clusters": clusters})
    cluster_df['id'] = cluster_df['id'].astype(str) 
    cluster_df['clusters'] = cluster_df['clusters'].astype(int)
    complete_subset = dataset.merge(cluster_df, left_on='node_id',right_on="id", how='left')

    
    # re-classification of outliers 
    cluster_sizes = complete_subset['clusters'].value_counts()
    valid_clusters = cluster_sizes[cluster_sizes == 1].index
    complete_subset.loc[complete_subset['clusters'].isin(valid_clusters), 'clusters'] = -1 # reassign those values to -1 for comparison with noises   

    clusters = complete_subset['clusters'].values # now all singletons are counted as a unique cluster
    num_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
    
    if num_clusters in range(2, dataset.shape[0]):   
        validation_metrics = compound_validity_index(complete_subset, matrix)
        
        # STORE THE RESULTS
        res = [grid, setting, dist, num_clusters] + validation_metrics
        
        return res
    else: 
        #print("no cluster identified")
        pass
```

# Agglomerative Hierarchical Clustering

```python
columns = ['polygon_id', 'buffer', 'dist', 'clusters',
           'clus_score',
           'ccc_score',
           'sil_score', 
           'dbi_score', 
           'cvs']
```

## ORIGINAL DATA

```python
# it may take a while
data_original = []

for polygon in data.polygon_id.unique():
    
    # retrieve subset of data points
    subset = data[data.polygon_id==polygon]
    
    # retrieve the corresponding distance matrix 
    matrix = dist_matrix_extraction("original", polygon)
    
    for dist in distance_parameter:
        
        # compute the clustering on street
        current_res = hac_clustering(dist, matrix, subset, polygon, "original")
        if current_res is not None:
            data_original.append(current_res)
    
```

```python
# turn into the results into dataframes
data_original = pd.DataFrame (data=data_original, columns=columns)

#check
data_original
```

## 100 M BUFFERED DATA

```python
# it may take a while
data_buffer100 = []

for polygon in data.polygon_id.unique():
    
    # retrieve subset for street
    matrix = dist_matrix_extraction("buffer_100", polygon)
    
    # we can restrieve directly the ids of data points inside the buffered polygons from the matrix
    current_nodes = [str(el) for el in matrix.index]
    subset = data[data.node_id.isin(current_nodes)]
    
    for dist in distance_parameter:
        
        # compute the clustering on street
        current_res = hac_clustering(dist, matrix, subset, polygon, "buffer_100")
        if current_res is not None:
            data_buffer100.append(current_res)
    
```

```python
# turn into the results into dataframes
data_buffer100 = pd.DataFrame (data=data_buffer100, columns=columns)
#check
data_buffer100
```

## 200 M BUFFER DATA

```python
# it may take a while
data_buffer200 = []

for polygon in data.polygon_id.unique():
    
    # retrieve subset for street
    matrix = dist_matrix_extraction("buffer_200", polygon)
    
    # we can restrieve directly the ids of data points inside the buffered polygons from the matrix
    current_nodes = [str(el) for el in matrix.index]
    subset = data[data.node_id.isin(current_nodes)]
    
    for dist in distance_parameter:
        
        # compute the clustering on street
        current_res = hac_clustering(dist, matrix, subset, polygon, "buffer_200")
        if current_res is not None:
            data_buffer200.append(current_res)
```

```python
# turn into the results into dataframes
data_buffer200 = pd.DataFrame (data=data_buffer200, columns=columns )

#check
data_buffer200
```

# result comparison

## Silhouette coefficient vs Cluster ratio

```python
for pol in sorted(data.polygon_id.unique()):
    
    # filtering based on polygon
    res_original = data_original[data_original['polygon_id'] == pol].copy()
    res_buffer100 = data_buffer100[data_buffer100['polygon_id'] == pol].copy()
    res_buffer200 = data_buffer200[data_buffer200['polygon_id'] == pol].copy()
    
    # objectives to maximize -> only two for the plot
    objectives = ['sil_score', 'clus_score']
    
    # computation of Pareto fronts separately
    pareto_mask_original = paretoset(res_original[objectives], sense=["max", "max"])
    pareto_mask_buffer100 = paretoset(res_buffer100[objectives], sense=["max", "max"])
    pareto_mask_buffer200 = paretoset(res_buffer200[objectives], sense=["max", "max"])
    
    # saving momentarily the results
    res_original_pareto = res_original[pareto_mask_original]
    res_buffer100_pareto = res_buffer100[pareto_mask_buffer100]
    res_buffer200_pareto = res_buffer200[pareto_mask_buffer200]
    
    # plotting for comparison
    fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
    
    datasets = [
        (res_original, res_original_pareto, "original shape", axes[0]),
        (res_buffer100, res_buffer100_pareto, "100 m buffer", axes[1]),
        (res_buffer200, res_buffer200_pareto, "200 m buffer", axes[2]),
    ]
    
    for full, pareto, title, ax in datasets:
        ax.scatter(full['sil_score'], full['clus_score'], color='lightgray', label='all runs', alpha=0.6)
        ax.scatter(pareto['sil_score'], pareto['clus_score'], color='red', label='Pareto front', s=60)
        ax.set_xlabel('Silhouette coefficient (normalized)')
        ax.set_xlim(0.0, 1.05)
        ax.set_ylim(0.0, 1.05)
        ax.legend(loc='lower left')
        ax.set_title(title)
        ax.grid(True)
    
    axes[0].set_ylabel('Cluster Ratio (normalized)')
    fig.suptitle(f"Pareto analysis for polygon n.{pol} on Hierarchical Agglomerative clustering results", fontsize=16)
    
    plt.tight_layout()
    plt.show()
```

## four-dimensional pareto analysis

```python
pareto_general_results = []

for pol in sorted(data.polygon_id.unique()):
    
    # filtering based on the polygon
    res_original = data_original[data_original['polygon_id'] == pol].copy()
    res_buffer100 = data_buffer100[data_buffer100['polygon_id'] == pol].copy()
    res_buffer200 = data_buffer200[data_buffer200['polygon_id'] == pol].copy()
    
    objectives = ['ccc_score', 'dbi_score', 'clus_score', 'sil_score']
    
    # Pareto computation
    pareto_mask_original = paretoset(res_original[objectives], sense=["max", "max", "max", "max"])
    pareto_mask_buffer100 = paretoset(res_buffer100[objectives], sense=["max", "max", "max", "max"])
    pareto_mask_buffer200 = paretoset(res_buffer200[objectives], sense=["max", "max", "max", "max"])
    
    # print see the results
    res_original_pareto = res_original[pareto_mask_original]
    res_buffer100_pareto = res_buffer100[pareto_mask_buffer100]
    res_buffer200_pareto = res_buffer200[pareto_mask_buffer200]
                                         
    # considering only the best ten results based on N (number of elements clustered)
    res_original_sorted = res_original_pareto.sort_values(by="cvs", ascending=False).head(10)
    res_buffer100_sorted = res_buffer100_pareto.sort_values(by="cvs", ascending=False).head(10)
    res_buffer200_sorted = res_buffer200_pareto.sort_values(by="cvs", ascending=False).head(10)
    
    
    # now we combine all the results obtained together
    best_results = pd.concat([res_original_sorted, res_buffer100_sorted, res_buffer200_sorted], ignore_index=True)
    
    # store them together                                    
    pareto_general_results.append(best_results)                                     
                                         
                                         
                                         
pareto_complete = pd.concat(pareto_general_results, ignore_index=True)  
                                         
# check
pareto_complete                                       
```

```python
# store the results for following computations
result_path = os.path.join(datasets_dir, "pareto_hac.csv")
pareto_complete.to_csv(result_path)

print(f"dataset stored in directory: {result_path}")
```

```python
cvs_map = [] 

for pol in sorted(data.polygon_id.unique()):
    
    # subsets
    res_original = data_original[data_original['polygon_id'] == pol].copy()
    res_buffer100 = data_buffer100[data_buffer100['polygon_id'] == pol].copy()
    res_buffer200 = data_buffer200[data_buffer200['polygon_id'] == pol].copy()
    
    pres_original = pareto_complete[(pareto_complete['polygon_id'] == pol) & (pareto_complete['buffer']=='original')].copy()
    pres_buffer100 = pareto_complete[(pareto_complete['polygon_id'] == pol) & (pareto_complete['buffer']=='buffer_100')].copy()
    pres_buffer200 = pareto_complete[(pareto_complete['polygon_id'] == pol) & (pareto_complete['buffer']=='buffer_200')].copy()
    
    # plot
    fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

    # --- original polygons
    sns.scatterplot(data=res_original, x='dist', y='cvs', ax=axes[0],
                    alpha=0.6, color='seagreen', label='all points')
    sns.scatterplot(data=pres_original, x='dist', y='cvs', ax=axes[0],
                    alpha=1.0, color='red', label='Pareto front')
    axes[0].invert_yaxis()
    axes[0].set_xlim([0.45, 2.6])
    axes[0].set_ylim([0.0, 1.05])
    axes[0].tick_params(axis='x', rotation=45)
    axes[0].xaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    axes[0].yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    axes[0].set_title("Original polygons")
    axes[0].set_xlabel("Cut distance threshold")
    axes[0].set_ylabel("Clustering Validity Score")

    # --- 100 m buffer
    sns.scatterplot(data=res_buffer100, x='dist', y='cvs', ax=axes[1],
                    alpha=0.6, color='seagreen', label='all points')
    sns.scatterplot(data=pres_buffer100, x='dist', y='cvs', ax=axes[1],
                    alpha=1.0, color='red', label='Pareto front')
    axes[1].invert_yaxis()
    axes[1].set_xlim([0.45, 2.6])
    axes[1].set_ylim([0.0, 1.05])
    axes[1].tick_params(axis='x', rotation=45)
    axes[1].xaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    axes[1].yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    axes[1].set_title("100 m buffer polygons")
    axes[1].set_xlabel("Cut distance threshold")

    # --- 200 m buffer
    sns.scatterplot(data=res_buffer200, x='dist', y='cvs', ax=axes[2],
                    alpha=0.6, color='seagreen', label='all points')
    sns.scatterplot(data=pres_buffer200, x='dist', y='cvs', ax=axes[2],
                    alpha=1.0, color='red', label='Pareto front')
    axes[2].invert_yaxis()
    axes[2].set_xlim([0.45, 2.6])
    axes[2].set_ylim([0.0, 1.05])
    axes[2].tick_params(axis='x', rotation=45)
    axes[2].xaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    axes[2].yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    axes[2].set_title("200 m buffer polygons")
    axes[2].set_xlabel("Cut distance threshold")

    # --- general settings
    plt.suptitle(f"Clustering validity score distribution for polygon n.{pol} with Hierarchical Agglomerative Clustering", fontsize=16)
    plt.tight_layout(rect=[0, 0.05, 1, 0.95])
    plt.show()
    
    # now we save the picture
    buf = io.BytesIO()
    fig.savefig(buf, format='png', bbox_inches='tight')
    buf.seek(0)
    img = Image.open(buf)
    cvs_map.append(img)
    plt.close(fig)

# global saving
total_height = sum(img.height for img in cvs_map)
max_width = max(img.width for img in cvs_map)

merged_image = Image.new('RGB', (max_width, total_height), (255, 255, 255))
y_offset = 0
for img in cvs_map:
    merged_image.paste(img, (0, y_offset))
    y_offset += img.height

merged_image.save("images/hac_cvs_pareto.png")
```
